useful built in function

```python
import numpy as np
scores=np.array([45,67,77,90,50])
print("maximum score:",scores.max())
print("minimum score:",scores.min())
print("average score:",scores.mean())
print("total score:",scores.sum())
print("sorted scores:",np.sort(scores))
```
